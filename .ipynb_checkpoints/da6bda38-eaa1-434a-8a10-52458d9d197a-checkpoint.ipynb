Данные из столбца days_employed относятся к механизму образования MNAR, мы должны выразить его с помощью возраста и на основании значения по медиане(покажу позднее почему так будет не совсем корректно), но он не нужен нам для анализа, так что можем не учитывать эти значения, удалим данный столбец. Такая ошибка могла возникнуть, к примеру, из-за того, что человек работал в чёрную и по нему нет никакой статистики, либо это ошибка при конвертации.

Докажу почему мы не можем корректно восстановить данные: во-первых, значения данного столбца зависят не только от dob_years, но и от медианы по общему трудовому стажу, выраженной по возрасту, а с ней у нас те ещё неполадки:

```python
negative = data.loc[data['days_employed'] < 0, 'days_employed'].count()
positive = data.loc[data['days_employed'] > 0, 'days_employed'].count()
negative_avg = data.loc[data['days_employed'] < 0, 'dob_years'].mean()
positive_avg = data.loc[data['days_employed'] > 0, 'dob_years'].mean()
negative_med = data.loc[data['days_employed'] < 0, 'dob_years'].median()
positive_med = data.loc[data['days_employed'] > 0, 'dob_years'].median()

print(f' Кол-во отрицательных значений в столбце days_employed: {negative}, среднее значение по возрасту равно {int(negative_avg)}, медианное значение по возрасту равно {int(negative_med)}')
display(data.loc[data['days_employed'] < 0, ['days_employed', 'dob_years']].sort_values(by='days_employed', ascending=False).head(10))

print(f' Кол-во положительных значений в столбце days_employed: {positive}, среднее значение по возрасту равно {int(positive_avg)}, медианное по возрасту значение равно {int(positive_med)}')
data.loc[data['days_employed'] > 0, ['days_employed', 'dob_years']].sort_values(by='days_employed', ascending=True).head(10)
```

Положительные значения в среднем это люди 60 лет, а отрицательные - 40(взял медиану и среднее для сравнения отклонения). Давайте создадим функцию, которая просчитает по первым значениям во сколько лет люди начали работать, отталкиваясь от того, что отрицательные значения - ошибка при заполнении, допустим это дни, просто умножим на -1, а положительные значения - это часы:

```python
def getting_started(row):
    if row['days_employed'] < 0:
        row['days_employed'] *= -1
        return row['dob_years']-(row['days_employed']/365)
    if row['days_employed'] > 0:
        return row['dob_years']-(row['days_employed']/24/365)
    

check = data.loc[data['days_employed'] > 0, ['days_employed', 'dob_years']].sort_values(by='days_employed', ascending=True)

check['age_starting'] = check.apply(getting_started, axis=1)
minors = check.loc[check['age_starting'] < 16, 'age_starting'].count()

print(f'Кол-во несовершеннолетних {minors}, они составляют {minors/positive:0.2%} от общего')
```

Вы только вдумайтесь, 37% некорректных данных - это очень много. Да, мы конечно можем добавить условие в функцию, по которому она будет отбирать только совершеннолетних, узнаем среднее по медиане и потом добавим + randint (чтобы значения были близкими к медиане, это называется модель стохастической линейной регрессии), но зачем нам такое искажение? Тем более, когда данный столбец не будет учавствовать в анализе. Но если бы он нам был необходим, следовало задать вопрос коллегам, которые делают выгрузку "А почему так? Не могли бы вы это исправить?". Данные по отрицательным значениям:

```python
check = data.loc[data['days_employed'] < 0, ['days_employed', 'dob_years']].sort_values(by='days_employed', ascending=False)
check['age_starting'] = check.apply(getting_started, axis=1)
minors = check.loc[check['age_starting'] < 16, 'age_starting'].count()

print(f'Кол-во несовершеннолетних {minors}, они составляют {minors/negative:0.2%} от общего')
```

Здесь всё выглядит получше, по первому взгляду скажет каждый, но давайте посмотрим на это с другой стороны: посчитаем людей, которые начали только работать после 30 лет, конечно, такая вероятность есть, но не более чем 5-7%, как мне кажется

```python
mature = check.loc[check['age_starting'] > 30, 'age_starting'].count()

print(f'Кол-во зрелых людей {mature}, они составляют {mature/negative:0.2%} от общего')
```

Как мы видим, данный столбец невозможно корректно восстановить ни по какому причастному к нему столбцу, поэтому он относиться к Missing Not At Random

С первым столбцом разобрались, перейдём к total_income. Здесь мы сталкиваемся с проблемой MNAR, но этот столбец мы можем исправить по income_type - выразим медиану по каждой категории и заменим вместо значения NaN. Почему медиану, а не среднее? Всё очень просто. К примеру, у нас есть зарплаты сотрудников 100, 100, 100, 100, 100, 100 и 500. Средним значением здесь будет 157, а медианным 100. С точки зрения аналитики - это некорректное решение.

```python
# Создадим DataFrame без NaN значений
not_nun_data = data.dropna()

#Создадим функцию, которая выведет нам среднее значение по столбцу total_income, опираясь на категориальные значения столбца income_type
def median_income(column):
    return int(column.median())

total_income_avg = not_nun_data.pivot_table(index='income_type',
                                           values='total_income',
                                           aggfunc={'count', median_income})
total_income_avg
```

Теперь займёмся устранением NaN значений

```python
unemployed = total_income_avg['median_income'][0]
decree = total_income_avg['median_income'][1]
civil_servant = total_income_avg['median_income'][2]
companion = total_income_avg['median_income'][3]
retiree = total_income_avg['median_income'][4]
businessman = total_income_avg['median_income'][5]
employee = total_income_avg['median_income'][6]
student = total_income_avg['median_income'][7]

data.loc[(data['total_income'].isna()) & (data['income_type'] == 'сотрудник'), 'total_income'] = employee
data.loc[(data['total_income'].isna()) & (data['income_type'] == 'компаньон'), 'total_income'] = companion
data.loc[(data['total_income'].isna()) & (data['income_type'] == 'пенсионер'), 'total_income'] = retiree
data.loc[(data['total_income'].isna()) & (data['income_type'] == 'госслужащий'), 'total_income'] = civil_servant
data.loc[(data['total_income'].isna()) & (data['income_type'] == 'предприниматель'), 'total_income'] = businessman

data.info()
```

Прекрасно, как итог: мы избавились от ненужного столбца days_employed и заполнили пропуски в total_income по медиане

### Категоризация данных

Выше я описывал для какого столбца нам понадобится категоризация - purpose, но также давайте оптимизируем аналилитику, добавим категории для столбцов total_income и children

```python
# Функция по созданию диапазонов значений заработка 
def total_income_gr(total_income):
    if total_income < 30001:
        return 'ЗП менее 30.000 рублей'
    if total_income < 50001:
        return 'ЗП от 30.000 до 50.000 рублей'
    if total_income < 200001:
        return 'ЗП от 50.000 до 200.000 рублей'
    if total_income < 1000001:
        return 'ЗП от 200.000 до 1.000.000 рублей'
    return 'ЗП более 1.000.000 рублей'

data['income_status'] = data['total_income'].apply(total_income_gr)

# Функция по группировке по кол-ву детей
def children_gr(children):
    if children == 0:
        return 'бездетная'
    if children < 3:
        return 'малодетная'
    return 'многодетная'

data['children_status'] = data['children'].apply(children_gr)


# Функция по распределению в категории на основании столбца purpose
def purpose_gr(purpose):
    if 'авто' in purpose:
        return 'операции с автомобилем'
    if 'свад' in purpose:
        return 'проведение свадьбы'
    if 'образ' in purpose:
        return 'получение образования'
    return 'операции с недвижимостью'

data['purpose_category'] = data['purpose'].apply(purpose_gr)

# Удалим столбец purpose, чтобы он не съедал память
del data['purpose']


display(data.head(10))
```

# Оптимизация 

```python
# Создадим копию DataFrame, чтобы в дальнейшем сравнить объёмы
data_before = data.copy()

# Проверим сколько памяти занимает данные файла
data.info(memory_usage='deep')

#Проверим сколько МБ занимает каждый подтип
for dtype in ['float','int','object']:
    selected_dtype = data.select_dtypes(include=[dtype])
    mean_usage_b = selected_dtype.memory_usage(deep=True).mean()
    mean_usage_mb = mean_usage_b / 1024 ** 2
    print("Average memory usage for {} columns: {:03.2f} MB".format(dtype,mean_usage_mb))
```

```python
#Уточним какой макимум и минимум может хранить подтип
print(np.iinfo('int64'))
print(np.iinfo('int32'))
```

У нас нет float значений, но такие значения оптимизируется ровно также, как и int. А начнём мы именно с него

```python
# Функция которая позволяет выводить объём в Мб
def mem_usage(pandas_obj):
    if isinstance(pandas_obj,pd.DataFrame):
        usage_b = pandas_obj.memory_usage(deep=True).sum()
    else: # исходим из предположения о том, что если это не DataFrame, то это Series
        usage_b = pandas_obj.memory_usage(deep=True)
    usage_mb = usage_b / 1024 ** 2 # преобразуем байты в мегабайты
    return "{:03.2f} MB".format(usage_mb)

# Оптимизация
data_int = data.select_dtypes(include=['int'])
converted_int = data_int.apply(pd.to_numeric,downcast='unsigned')

print(mem_usage(data_int))
print(mem_usage(converted_int))

# Выведем таблицу до и после
compare_ints = pd.concat([data_int.dtypes,converted_int.dtypes],axis=1)
compare_ints.columns = ['before','after']
compare_ints.apply(pd.Series.value_counts)
```

Это всё прекрасно, но тип данных object всегда будет весить больше, чем int (единичный экземпляр каждого). Давайте займёмся им

```python
print(mem_usage(data))

for col in data.columns:
    num_unique_values = len(data[col].unique())
    num_total_values = len(data[col])
    if num_unique_values / num_total_values < 0.5:
        data.loc[:,col] = data[col].astype('category')
    else:
        data.loc[:,col] = data[col]
        

print(mem_usage(data))

data_info = pd.concat([data_before.dtypes,data.dtypes],axis=1)
data_info.columns = ['before','after']
data_info.apply(pd.Series.value_counts)
print(data_info)
data.info(memory_usage='deep')
```

Отлично, оптимизация файла прошла успешно, выигрыш составляет 98%

### Есть ли зависимость между количеством детей и возвратом кредита в срок?

```python
# Напишем функцию, которая будет вычислять процент, чтобы каждый раз не заниматься писаниной
def debt_precent(column):
    return '{:.2f}'.format(column.sum()/column.count() * 100) + '%'
```

```python
pivot_children = data_before.pivot_table(index={'children_status', 'children'},
                                 values='debt',
                                 aggfunc={'count', debt_precent})
pivot_children
```

Для начала удалим children == 5, тк она сбивает анализ из-за малого кол-ва данных, потом выведем анализ по children_status

```python
data_before = data_before.loc[data_before['children'] < 5]

pivot_children = data_before.pivot_table(index='children_status',
                                 values='debt',
                                 aggfunc={'count', debt_precent})


pivot_children = pivot_children['debt_precent'].sort_values(ascending=False)
pivot_children
```
