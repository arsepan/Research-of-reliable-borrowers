```python
# –°–æ–∑–¥–∞–¥–∏–º DataFrame –±–µ–∑ NaN –∑–Ω–∞—á–µ–Ω–∏–π
not_nun_data = data.dropna()

#–°–æ–∑–¥–∞–¥–∏–º —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –≤—ã–≤–µ–¥–µ—Ç –Ω–∞–º —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –ø–æ —Å—Ç–æ–ª–±—Ü—É total_income, –æ–ø–∏—Ä–∞—è—Å—å –Ω–∞ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è —Å—Ç–æ–ª–±—Ü–∞ income_type
def median_income(column):
    return int(column.median())

total_income_avg = not_nun_data.pivot_table(index='income_type',
                                           values='total_income',
                                           aggfunc={'count', median_income})
total_income_avg
```

```python
unemployed = total_income_avg['median_income'][0]
decree = total_income_avg['median_income'][1]
civil_servant = total_income_avg['median_income'][2]
companion = total_income_avg['median_income'][3]
retiree = total_income_avg['median_income'][4]
businessman = total_income_avg['median_income'][5]
employee = total_income_avg['median_income'][6]
student = total_income_avg['median_income'][7]

data.loc[(data['total_income'].isna()) & (data['income_type'] == '—Å–æ—Ç—Ä—É–¥–Ω–∏–∫'), 'total_income'] = employee
data.loc[(data['total_income'].isna()) & (data['income_type'] == '–∫–æ–º–ø–∞–Ω—å–æ–Ω'), 'total_income'] = companion
data.loc[(data['total_income'].isna()) & (data['income_type'] == '–ø–µ–Ω—Å–∏–æ–Ω–µ—Ä'), 'total_income'] = retiree
data.loc[(data['total_income'].isna()) & (data['income_type'] == '–≥–æ—Å—Å–ª—É–∂–∞—â–∏–π'), 'total_income'] = civil_servant
data.loc[(data['total_income'].isna()) & (data['income_type'] == '–ø—Ä–µ–¥–ø—Ä–∏–Ω–∏–º–∞—Ç–µ–ª—å'), 'total_income'] = businessman

data.info()
```

–í—ã—à–µ —è –æ–ø–∏—Å—ã–≤–∞–ª –¥–ª—è –∫–∞–∫–æ–≥–æ —Å—Ç–æ–ª–±—Ü–∞ –Ω–∞–º –ø–æ–Ω–∞–¥–æ–±–∏—Ç—Å—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∑–∞—Ü–∏—è - purpose, –Ω–æ —Ç–∞–∫–∂–µ –¥–∞–≤–∞–π—Ç–µ –æ–ø—Ç–∏–º–∏–∑–∏—Ä—É–µ–º –∞–Ω–∞–ª–∏–ª–∏—Ç–∏–∫—É, –¥–æ–±–∞–≤–∏–º –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –¥–ª—è —Å—Ç–æ–ª–±—Ü–æ–≤ total_income –∏ children

```python
# –§—É–Ω–∫—Ü–∏—è –ø–æ —Å–æ–∑–¥–∞–Ω–∏—é –¥–∏–∞–ø–∞–∑–æ–Ω–æ–≤ –∑–Ω–∞—á–µ–Ω–∏–π –∑–∞—Ä–∞–±–æ—Ç–∫–∞ 
def total_income_gr(total_income):
    if total_income < 30001:
        return '–ó–ü –º–µ–Ω–µ–µ 30.000 —Ä—É–±–ª–µ–π'
    if total_income < 50001:
        return '–ó–ü –æ—Ç 30.000 –¥–æ 50.000 —Ä—É–±–ª–µ–π'
    if total_income < 200001:
        return '–ó–ü –æ—Ç 50.000 –¥–æ 200.000 —Ä—É–±–ª–µ–π'
    if total_income < 1000001:
        return '–ó–ü –æ—Ç 200.000 –¥–æ 1.000.000 —Ä—É–±–ª–µ–π'
    return '–ó–ü –±–æ–ª–µ–µ 1.000.000 —Ä—É–±–ª–µ–π'

data['income_status'] = data['total_income'].apply(total_income_gr)

# –§—É–Ω–∫—Ü–∏—è –ø–æ –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫–µ –ø–æ –∫–æ–ª-–≤—É –¥–µ—Ç–µ–π
def children_gr(children):
    if children == 0:
        return '–±–µ–∑–¥–µ—Ç–Ω–∞—è'
    if children < 3:
        return '–º–∞–ª–æ–¥–µ—Ç–Ω–∞—è'
    return '–º–Ω–æ–≥–æ–¥–µ—Ç–Ω–∞—è'

data['children_status'] = data['children'].apply(children_gr)


# –§—É–Ω–∫—Ü–∏—è –ø–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—é –≤ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ –Ω–∞ –æ—Å–Ω–æ–≤–∞–Ω–∏–∏ —Å—Ç–æ–ª–±—Ü–∞ purpose
def purpose_gr(purpose):
    if '–∞–≤—Ç–æ' in purpose:
        return '–æ–ø–µ—Ä–∞—Ü–∏–∏ —Å –∞–≤—Ç–æ–º–æ–±–∏–ª–µ–º'
    if '—Å–≤–∞–¥' in purpose:
        return '–ø—Ä–æ–≤–µ–¥–µ–Ω–∏–µ —Å–≤–∞–¥—å–±—ã'
    if '–æ–±—Ä–∞–∑' in purpose:
        return '–ø–æ–ª—É—á–µ–Ω–∏–µ –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è'
    return '–æ–ø–µ—Ä–∞—Ü–∏–∏ —Å –Ω–µ–¥–≤–∏–∂–∏–º–æ—Å—Ç—å—é'

data['purpose_category'] = data['purpose'].apply(purpose_gr)

# –£–¥–∞–ª–∏–º —Å—Ç–æ–ª–±–µ—Ü purpose, —á—Ç–æ–±—ã –æ–Ω –Ω–µ —Å—ä–µ–¥–∞–ª –ø–∞–º—è—Ç—å
del data['purpose']


display(data.head(10))
```

# –û–ø—Ç–∏–º–∏–∑–∞—Ü–∏—è 

```python
# –°–æ–∑–¥–∞–¥–∏–º –∫–æ–ø–∏—é DataFrame, —á—Ç–æ–±—ã –≤ –¥–∞–ª—å–Ω–µ–π—à–µ–º —Å—Ä–∞–≤–Ω–∏—Ç—å –æ–±—ä—ë–º—ã
data_before = data.copy()

# –ü—Ä–æ–≤–µ—Ä–∏–º —Å–∫–æ–ª—å–∫–æ –ø–∞–º—è—Ç–∏ –∑–∞–Ω–∏–º–∞–µ—Ç –¥–∞–Ω–Ω—ã–µ —Ñ–∞–π–ª–∞
data.info(memory_usage='deep')

#–ü—Ä–æ–≤–µ—Ä–∏–º —Å–∫–æ–ª—å–∫–æ –ú–ë –∑–∞–Ω–∏–º–∞–µ—Ç –∫–∞–∂–¥—ã–π –ø–æ–¥—Ç–∏–ø
for dtype in ['float','int','object']:
    selected_dtype = data.select_dtypes(include=[dtype])
    mean_usage_b = selected_dtype.memory_usage(deep=True).mean()
    mean_usage_mb = mean_usage_b / 1024 ** 2
    print("Average memory usage for {} columns: {:03.2f} MB".format(dtype,mean_usage_mb))
```

–£ –Ω–∞—Å –Ω–µ—Ç float –∑–Ω–∞—á–µ–Ω–∏–π, –Ω–æ —Ç–∞–∫–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è –æ–ø—Ç–∏–º–∏–∑–∏—Ä—É–µ—Ç—Å—è —Ä–æ–≤–Ω–æ —Ç–∞–∫–∂–µ, –∫–∞–∫ –∏ int. –ê –Ω–∞—á–Ω—ë–º –º—ã –∏–º–µ–Ω–Ω–æ —Å –Ω–µ–≥–æ

```python
# –§—É–Ω–∫—Ü–∏—è –∫–æ—Ç–æ—Ä–∞—è –ø–æ–∑–≤–æ–ª—è–µ—Ç –≤—ã–≤–æ–¥–∏—Ç—å –æ–±—ä—ë–º –≤ –ú–±
def mem_usage(pandas_obj):
    if isinstance(pandas_obj,pd.DataFrame):
        usage_b = pandas_obj.memory_usage(deep=True).sum()
    else: # –∏—Å—Ö–æ–¥–∏–º –∏–∑ –ø—Ä–µ–¥–ø–æ–ª–æ–∂–µ–Ω–∏—è –æ —Ç–æ–º, —á—Ç–æ –µ—Å–ª–∏ —ç—Ç–æ –Ω–µ DataFrame, —Ç–æ —ç—Ç–æ Series
        usage_b = pandas_obj.memory_usage(deep=True)
    usage_mb = usage_b / 1024 ** 2 # –ø—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –±–∞–π—Ç—ã –≤ –º–µ–≥–∞–±–∞–π—Ç—ã
    return "{:03.2f} MB".format(usage_mb)

# –û–ø—Ç–∏–º–∏–∑–∞—Ü–∏—è
data_int = data.select_dtypes(include=['int'])
converted_int = data_int.apply(pd.to_numeric,downcast='unsigned')

print(mem_usage(data_int))
print(mem_usage(converted_int))

# –í—ã–≤–µ–¥–µ–º —Ç–∞–±–ª–∏—Ü—É –¥–æ –∏ –ø–æ—Å–ª–µ
compare_ints = pd.concat([data_int.dtypes,converted_int.dtypes],axis=1)
compare_ints.columns = ['before','after']
compare_ints.apply(pd.Series.value_counts)
```
